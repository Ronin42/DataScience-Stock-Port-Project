```python
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome(r'D:\Python lecture\chromedriver.exe')
driver.get('https://www.set.or.th/th/market/index/set100/overview')
data = driver.page_source
data_df = pd.read_html(data)[2]
clean_column =[]
for x in data_df.columns:
    clean_column.append(x.replace(' (Click to sort Ascending)',''))
data_df.columns = clean_column
data_df = data_df.set_index('หลักทรัพย์ ')
data_df
```

```python
import time
```

```python
all_stock_dict = {}
driver = webdriver.Chrome(r'D:\Python lecture\chromedriver.exe')


for stock in data_df.index:
    stock = stock.split()[0]
    url = 'https://www.set.or.th/th/market/product/stock/quote/'+stock+'/financial-statement/company-highlights'
    driver.get(url)
    time.sleep(1)
    stock_data = driver.page_source
    a_df = pd.read_html(stock_data)[0]
    b_df = pd.read_html(stock_data)[1]
    all_stock_dict[stock] = [a_df,b_df]
    
```

```python
def get_stock_data(stock,driver,all_stock_dict):
    stock = stock.split()[0]
    url = 'https://www.set.or.th/th/market/product/stock/quote/'+stock+'/financial-statement/company-highlights'
    driver.get(url)
    stock_data = driver.page_source
    a_df = pd.read_html(stock_data)[0]
    a_df = a_df.rename_axis(stock)
    b_df = pd.read_html(stock_data)[1]
    b_df = b_df.rename_axis(stock)
    all_stock_dict[stock] = [a_df,b_df]
    return driver,all_stock_dict
```
