# dfdebt2019 

```python
# นำข้อมูลจาก balancsheet แต่ละปีมาคำนวณแล้วสร้างตารางใหม่

# currentRatio =  Total Current Assets/Total Current Liabilities
# Current Ratioหรืออัตราส่วนเงินทุนหมุนเวียน เป็นอัตราส่วนที่บอกถึงความสามารถในการชำระหนี้สินระยะสั้นว่าเป็นอย่างไร
# หาก Current Ratio นั้นมีค่าน้อยกว่า 1 แสดงว่าบริษัทมีความเสี่ยงที่จะขาดสภาพคล่องในระยะสั้น

# debtToAssets = Total Liab/total asset
# อัตราส่วนนี้สามารถสะท้อนถึงความมั่นคงทางการเงินของบริษัท ยิ่งอัตราส่วนสูงเท่าใด ก็จะยิ่งทำให้ความเสี่ยงในการลงทุนในบริษัทนั้นสูงขึ้นด้วย

# debtToEquity = Total Liab/TotalStockholderEquity
# ค่า DE Ratio ยิ่งสูง แสดงว่า บริษัทนั้นยิ่งมีความเสี่ยงทางการเงินเพิ่มขึ้นด้วย
# จากการที่มีหนี้สินสูงเกินไป ซึ่งทำให้บริษัทมีภาระดอกเบี้ยที่ต้องจ่ายสูงตามไปด้วย
# จนอาจส่งผลต่อฐานะทางการเงินและความสามารถในการชำระหนี้ของบริษัทได้

dfdebt2019 = pd.DataFrame(index=close2019.keys(),columns=['currentRatio','debtToAssets','debtToEquity'])
```

```python
#dfdebt2019 currentRatio
selectcol = 'TotalCurrentAssets'
selectcol2 = 'TotalCurrentLiabilities'

for stockname in dfdebt2019.index:
    try:
        thisTotalCurrentAssets = float(dfbalanc2019[selectcol][stockname])
        thisTotalCurrentLiabilities = float(dfbalanc2019[selectcol2][stockname])

        dfdebt2019.loc[stockname,['currentRatio']] = thisTotalCurrentAssets/thisTotalCurrentLiabilities
    except KeyError:
        dfdebt2019.loc[stockname,['currentRatio']] = np.nan
```

```python
#dfdebt2019 debtToAssets
selectcol = 'TotalLiab'
selectcol2 = 'TotalAssets'

for stockname in dfdebt2019.index:
    try:
        thisTotalLiab = float(dfbalanc2019[selectcol][stockname])
        thisTotalAssets = float(dfbalanc2019[selectcol2][stockname])

        dfdebt2019.loc[stockname,['debtToAssets']] = thisTotalLiab/thisTotalAssets
    except KeyError:
        dfdebt2019.loc[stockname,['debtToAssets']] = np.nan
```

```python
#dfdebt2019 debtToEquity
selectcol = 'TotalLiab'
selectcol2 = 'TotalStockholderEquity'

for stockname in dfdebt2019.index:
    try:
        thisTotalLiab = float(dfbalanc2019[selectcol][stockname])
        thisTotalEquity = float(dfbalanc2019[selectcol2][stockname])

        dfdebt2019.loc[stockname,['debtToEquity']] = thisTotalLiab/thisTotalEquity
    except KeyError:
        dfdebt2019.loc[stockname,['debtToEquity']] = np.nan
```
